```python
#...Image Extraction from Numpy Arrays...#

#Import Statements
import numpy
import matplotlib.pyplot as plt

#Loading the Numpy Arrays
trainX = numpy.load('/Users/mysterymachine/Desktop/Project3/Data/tinyX.npy') # this should have shape (26344, 3, 64, 64)
trainY = numpy.load('/Users/mysterymachine/Desktop/Project3/Data/tinyY.npy') 
testX = numpy.load('/Users/mysterymachine/Desktop/Project3/Data/tinyX_test.npy') # (6600, 3, 64, 64)
```

```python
#Naming Training Dataset
Train_Images = r'/Users/mysterymachine/Desktop/Project3/Jpeg/Train_Image/Img{0:02d}.jpg'

#Running through the entire (Training) Numpy array
for i in range(len(trainX)):
    
    #To visualize the image
    fig = plt.imshow(trainX[i].transpose(2,1,0))
    
    #To cut off axis from the plot
    plt.axis('off')
    fig.axes.get_xaxis().set_visible(False)
    fig.axes.get_yaxis().set_visible(False)
    
    #Saving the figure - Format (PNG - Loseless Image), Removing White spaces and focussing on image
    #Extracted Image will be (224 x 224)
    plt.savefig(Train_Images.format(i), format='jpeg', bbox_inches='tight', pad_inches=0)
    
    #Clearing the frame for the next image
    plt.clf()
```
